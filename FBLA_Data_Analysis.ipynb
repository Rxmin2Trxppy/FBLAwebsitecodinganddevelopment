```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/FBLADataAnalysis/data/airquality.csv")

# Cleaning the data

# Drop columns with no meaningful data (e.g., 'Message')
data_cleaned = data.drop(columns=['Message'])

# Check for missing values and handle them
missing_summary = data_cleaned.isnull().sum()

# Convert date columns to datetime format
data_cleaned['Start_Date'] = pd.to_datetime(data_cleaned['Start_Date'], errors='coerce')

print(data_cleaned["Name"].unique())


# Display the cleaned dataset structure and missing data summary
pm25_data = data_cleaned[data_cleaned['Name'].str.contains('Fine particles \(PM 2.5\)', na=False)]
# Group by year and calculate the average PM2.5 levels
pm25_data['Year'] = pm25_data['Start_Date'].dt.year
pm25_trends = pm25_data.groupby('Year')['Data Value'].mean()

# Plot PM2.5 trends over time
plt.figure(figsize=(10, 6))
pm25_trends.plot(marker='o', color='teal', title='Average PM2.5 Levels Over Time')
plt.xlabel('Year')
plt.ylabel('PM2.5 (g/m)')
plt.grid(True)
plt.show()
```

```python
no2_data = data_cleaned[data_cleaned['Name'].str.contains('NO2', na=False)]
no2_data['Year'] = no2_data['Start_Date'].dt.year
no2_trends = no2_data.groupby('Year')['Data Value'].mean()

plt.figure(figsize=(12, 6))
plt.plot(no2_trends.index, no2_trends.values, label='NO2 Levels', marker='o', color='blue')
plt.title('Trends in NO2 Levels Over Time')
plt.xlabel('Year')
plt.ylabel('Concentration (ppm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
so2_data = data_cleaned[data_cleaned['Name'].str.contains('SO2', na=False)]
so2_data['Year'] = so2_data['Start_Date'].dt.year
so2_trends = so2_data.groupby('Year')['Data Value'].mean()

plt.figure(figsize=(12, 6))
plt.plot(so2_trends.index, so2_trends.values, label='SO2 Levels', marker='o', color='blue')
plt.title('Trends in SO2 Levels Over Time')
plt.xlabel('Year')
plt.ylabel('Concentration (ppm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
ozone_data = data_cleaned[data_cleaned['Name'].str.contains('O3', na=False)]
ozone_data['Year'] = ozone_data['Start_Date'].dt.year
ozone_trends = ozone_data.groupby('Year')['Data Value'].mean()

plt.figure(figsize=(12, 6))
plt.plot(ozone_trends.index, ozone_trends.values, label='Ozone Levels', marker='o', color='blue')
plt.title('Trends in Ozone Levels Over Time')
plt.xlabel('Year')
plt.ylabel('Concentration (ppb)')
plt.legend()
plt.grid(True)
plt.show()
```
