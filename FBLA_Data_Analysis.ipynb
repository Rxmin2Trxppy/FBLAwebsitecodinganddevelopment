```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/FBLADataAnalysis/data/airquality.csv")

# Cleaning the data

# Drop columns with no meaningful data (e.g., 'Message')
data_cleaned = data.drop(columns=['Message'])

# Check for missing values and handle them
missing_summary = data_cleaned.isnull().sum()

# Convert date columns to datetime format
data_cleaned['Start_Date'] = pd.to_datetime(data_cleaned['Start_Date'], errors='coerce')

print(data_cleaned["Name"].unique())
```

```python

#PM2.5 Levels
datatable = data_cleaned[data_cleaned['Name'].str.contains('Fine particles \(PM 2.5\)', na=False)]
datatable['Year'] = datatable['Start_Date'].dt.year
trends = datatable.groupby('Year')['Data Value'].mean()
plt.figure(figsize=(10, 6))
trends.plot(marker='o', color='teal', title='Average PM2.5 Levels Over Time')
plt.xlabel('Year')
plt.ylabel('PM2.5 (µg/m³)')
plt.grid(True)
plt.show()
```
